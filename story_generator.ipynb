# Step1:

Setting up the necessary imports and configuration of the model



```python
from transformers import pipeline, set_seed
import gradio as gr

set_seed(42)

MODEL_NAME = "gpt2"
DEFAULT_MAX_LENGTH = 150
DEFAULT_TEMPERATURE = 0.8
```

```python
try:
    generator = pipeline('text-generation', model=MODEL_NAME)
    print(f"GPT-2 model loaded successfully for Gradio.")
except Exception as e:
    print(f"Error loading model: {e}")
    generator = None
```

```python
def generate_story_web(prompt, max_len, temp):
    if not generator:
        return "Model not loaded. Please check the setup cell."

    # Clamp the parameters to prevent extreme values
    max_len = int(max(50, min(max_len, 500)))
    temp = float(max(0.1, min(temp, 1.5)))

    print(f"Generating story with Max Length: {max_len}, Temperature: {temp}")

    result = generator(
        prompt,
        max_length=max_len,
        temperature=temp,
        do_sample=True,
        pad_token_id=generator.tokenizer.eos_token_id,
        num_return_sequences=1
    )

    # Return the generated text string
    generated_text = result[0]['generated_text']
    return generated_text
```

```python
input_components = [
    gr.Textbox(
        lines=2,
        label="Story Starter Prompt",
        value="A curious child found a glowing key under a cherry tree.",
        placeholder="Start your story here..."
    ),
    gr.Slider(
        minimum=50,
        maximum=500,
        step=10,
        value=DEFAULT_MAX_LENGTH,
        label="Max Story Length (Tokens)",
        info="Controls the maximum length of the generated text."
    ),
    gr.Slider(
        minimum=0.1,
        maximum=1.5,
        step=0.05,
        value=DEFAULT_TEMPERATURE,
        label="Creativity/Randomness (Temperature)",
        info="Higher values produce more random/creative text."
    ),
]
```

```python
output_component = gr.Textbox(
    label="✨ Generated Story/Poem ✨",
    lines=10
)

# 3. Create and launch the Gradio Interface
iface = gr.Interface(
    fn=generate_story_web,
    inputs=input_components,
    outputs=output_component,
    title="🤖 Creative AI Story Generator (GPT-2)",
    description="Interact with the GPT-2 model to generate stories by adjusting the prompt and parameters."
)

# Launch the app! It will appear right here in your notebook.
iface.launch()
```
